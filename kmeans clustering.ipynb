```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import pandas as pd
a= pd.read_csv("data.csv")
print(a)
x=a.iloc[:,:].values
print(x)

#no traing test required
from sklearn.cluster import KMeans
kmeans=KMeans(init='k-means++',n_clusters=3,n_init=10,random_state=0).fit(x)
p=kmeans.predict(x)

import matplotlib.pyplot as plt
colors=list(map(lambda x:'#3b4cc0' if x==1 else '#000000',p))
plt.scatter(x[:,0],x[:,1],c=colors,marker='+',picker=True)
 
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
 
a=load_iris()
x=a.data
y=a.target
 
# print(x)
 
#no traing test required
from sklearn.cluster import KMeans
kmeans=KMeans(init='k-means++',n_clusters=2,n_init=10,random_state=0).fit(x)
p=kmeans.predict(x)
print(p)
import matplotlib.pyplot as plt
 
colors=list(map(lambda x:'#3b4cc0' if x==1 else '#000000',p))
plt.scatter(x[:,0],x[:,1],c=colors,marker='*',picker=True)
 
plt.show()
```
